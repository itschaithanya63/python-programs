```python
### def train_reservation():
    # Initialize 20 seats to 0 (0 = available, 1 = booked)
    seats = [0] * 20
    
    while True:
        # Displaying the Menu
        print("\n--- Train Seat Reservation System ---")
        print("1. Book Seat")
        print("2. Cancel Seat")
        print("3. Show Seats")
        print("4. Booking Percentage")
        print("5. Exit")
        
        # User input (assumes user will enter an integer)
        choice = int(input("Enter your choice (1-5): "))
        
        # 1. Book a seat
        if choice == 1:
            seat_num = int(input("Enter seat number to book (1-20): "))
            if 1 <= seat_num <= 20:
                if seats[seat_num - 1] == 0:
                    seats[seat_num - 1] = 1
                    print(f"Success: Seat {seat_num} has been booked!")
                else:
                    print(f"Error: Seat {seat_num} is already booked.")
            else:
                print("Invalid seat number. Please choose between 1 and 20.")
                
        # 2. Cancel a seat
        elif choice == 2:
            seat_num = int(input("Enter seat number to cancel (1-20): "))
            if 1 <= seat_num <= 20:
                if seats[seat_num - 1] == 1:
                    seats[seat_num - 1] = 0
                    print(f"Success: Booking for seat {seat_num} has been cancelled!")
                else:
                    print(f"Error: Seat {seat_num} is not currently booked.")
            else:
                print("Invalid seat number. Please choose between 1 and 20.")
                
        # 3. Display available seats
        elif choice == 3:
            print("\nAvailable Seats:")
            available_seats = []
            for i in range(20):
                if seats[i] == 0:
                    # Append actual seat number (index + 1)
                    available_seats.append(str(i + 1))
                    
            if available_seats:
                print(", ".join(available_seats))
            else:
                print("No seats available. The train is fully booked.")
                
        # 4. Display booking percentage
        elif choice == 4:
            booked_count = seats.count(1)
            percentage = (booked_count / 20) * 100
            print(f"\nCurrent Booking Percentage: {percentage}%")
            
        # 5. Exit system
        elif choice == 5:
            print("Exiting system. Have a safe journey!")
            break
            
        # Handle invalid menu choices
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

# Run the program
if __name__ == "__main__":
    train_reservation()
```

```python
def track_hashtags(hashtags):
    if not hashtags:
        print("No hashtags to analyze.")
        return

    unique_hashtags = list(set(hashtags))

    frequency = {}
    for tag in unique_hashtags:
        frequency[tag] = hashtags.count(tag)

    most_repeated = max(frequency, key=frequency.get)

    print("--- Social Media Hashtag Tracker ---")
    print(f"Original List: {hashtags}\n")
    print(f"Unique Hashtags (No Duplicates): {unique_hashtags}")
    print("Frequency of each hashtag:")
    for tag, count in frequency.items():
        print(f"  {tag}: {count} times")
    print(f"\nMost Repeated Hashtag: {most_repeated}")

sample_hashtags = ["#fun", "#python", "#fun", "#code", "#python"]

track_hashtags(sample_hashtags)
```

```python
def pattern_compression(input_list):
    if not input_list:
        return []

    compressed_output = []
    current_element = input_list[0]
    count = 0

    for item in input_list:
        if item == current_element:
            count += 1
        else:
            compressed_output.append((current_element, count))
            current_element = item
            count = 1
            
    compressed_output.append((current_element, count))
    
    return compressed_output

data = [1, 1, 1, 2, 2, 3, 3, 3]
result = pattern_compression(data)

print(f"Input : {data}")
print(f"Output: {result}")
```

```python
def shopping_cart_system():
    cart = []
    prices = {"apple": 10, "banana": 5, "milk": 20, "bread": 15}
    
    print("--- Mini Shopping Cart ---")
    print("Available items: apple, banana, milk, bread")

    while True:
        print("\n1. Add Item | 2. Remove Item | 3. Search Item | 4. Total Bill | 5. Show Duplicates | 6. Exit")
        choice = input("Select an option: ")

        if choice == '1':
            item = input("Enter item to add: ").lower()
            cart.append(item)
            print(f"'{item}' added to cart.")

        elif choice == '2':
            item = input("Enter item to remove: ").lower()
            if item in cart:
                cart.remove(item)
                print(f"'{item}' removed from cart.")
            else:
                print("Item not found in cart.")

        elif choice == '3':
            item = input("Search for: ").lower()
            count = 0
            for i in cart:
                if i == item:
                    count += 1
            if count > 0:
                print(f"'{item}' is in your cart (Quantity: {count}).")
            else:
                print(f"'{item}' is not in your cart.")

        elif choice == '4':
            total = 0
            for item in cart:
                total += prices.get(item, 0) 
            print(f"Total Bill: ${total}")

        elif choice == '5':
            duplicates = []
            seen = []
            for item in cart:
                if item in seen and item not in duplicates:
                    duplicates.append(item)
                seen.append(item)
            
            if duplicates:
                print(f"Duplicate items added: {duplicates}")
            else:
                print("No duplicates found.")

        elif choice == '6':
            print("Thank you for shopping!")
            break
        else:
            print("Invalid choice, try again.")


shopping_cart_system()
```
