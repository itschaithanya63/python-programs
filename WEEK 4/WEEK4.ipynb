```python
print("===LIST EXAMPLE===")
fruits=["apple","banana","cherry","apple"] #duplicates allowed
print("Original List",fruits)
```

```python
fruits.append("pomogrante")
fruits.remove("apple")
print(fruits)
```

```python
print(fruits[0]) #can access by index as it is ordered
```

```python
upper_fruits=[fruit.upper() for fruit in fruits]
print("Uppercase Fruits:",upper_fruits)
```

```python
print("\n===TUPLE EXAMPLE===")
coordinates=(10.5,20.3,10)
print("Tuple:",coordinates)
print(coordinates[0])
print(coordinates[-1])
```

```python
#unpacking
x,y,z=coordinates
print(f"Unpacked:x={x},y={y},z={z}")
```
