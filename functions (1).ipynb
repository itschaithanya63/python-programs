```python
#Simple function 
def greet():
    print("Hello,World!")
greet()
```

```python
#Function with parameters 
def add(a,b):
     return a+b
print(add(3,5))
```

```python
#Function with default argument
def power(base,exp=2):
    return base ** exp
print(power(4)) #default squares
print(power(3,4))
```

```python
#Function returning multiple values
def stats(numbers):
     return min(numbers),max(numbers),sum(numbers)/len(numbers)
print(stats([2,4,6,8]))
```

```python
#Lamba function #in line function
square=lambda x: x**2
print(square(9))
```

```python
#Basic callback
def process(data, callback):
    result=data*2
    callback(result)

def show_output(value):
    print("Processed:",value)

process(7,show_output)
```

```python
#callback with lamba
process(10, lambda x:print("Lambda result:",x))
```
