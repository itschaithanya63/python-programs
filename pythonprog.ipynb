```python
year=int(input("Enter year"))
if year%4==0 and year%100!=0:
    print("Leap year")
else:
    print("Not a leap year")
```

```python
c=str(input("Enter a character"))
if c in['A','E','I','O','U'] or c in['a','e','i','o','u']:
    print("Vowel")
else :
    print("Consonant")
```

```python
m1=float(input("Enter subject1 marks"))
m2=float(input("Enter subject2 marks"))
m3=float(input("Enter subject3 marks"))
tot=m1+m2+m3
avg=tot/3
if avg>=90:
    print("Grade=A")
elif avg>=80:
    print("Grade=B")
elif avg>=70:
    print("Grade=C")
elif avg>=60:
    print("Garde=D")
elif avg>=50:
    print("Grade=E")
else:
    print("FAIL!")
```

```python
num=int(input("Enter a number"))
if num%5==0 and num%11==0:
    print("Divisible by both 5 and 11")
else:
    print("Not divisible by both 5 and 11")
```

```python
x=input()
if x.isalpha():
    print("Alphabet")
elif x.isdigit():
    print("Digit")
else:
    print("Special Character")
```

```python
num=int(input("Enter a number"))
fact=1
for i in range(1,num+1):
    fact*=i;
print("Factorial=",fact)
```
